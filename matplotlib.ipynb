```python
!pip install datasets
import pandas as pd
from datasets import load_dataset

#load data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

#clean up
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
df
```

```python
import matplotlib.pyplot as plt
import numpy as np

x=np.arange(0,5,0.1)
y=np.sin(x)
plt.plot(x,y)
plt.xlabel('x')
plt.ylabel('sin(x)')
plt.show()
```

```python
#plot by job_posted_date
job_dated = df['job_posted_date'].value_counts()
job_dated = job_dated.sort_index()
plt.plot(job_dated.index, job_dated.values)
plt.xlabel('Date')
plt.ylabel('Number of jobs posted')
plt.show()
```

```python
# Extract the month from the 'job_posted_date' column and create a new column 'job_posted_month'
df['job_posted_month'] = df['job_posted_date'].dt.month

# Count the number of jobs posted in each month and store the result in 'job_posted_month'
job_posted_month = df['job_posted_month'].value_counts()

# Sort the counts by the month to ensure the months are in chronological order
job_posted_month = job_posted_month.sort_index()

# Plot the number of jobs posted per month
plt.plot(job_posted_month.index, job_posted_month.values)

# Label the x-axis as 'Month'
plt.xlabel('Month')

# Label the y-axis as 'Number of jobs posted'
plt.ylabel('Number of jobs posted')

# Display the plot
plt.show()
```

```python
job_titled=df['job_title_short'].value_counts()
job_titled = job_titled.sort_index()
plt.bar(job_titled.index,job_titled)
plt.xlabel('Job Title')
plt.xticks(rotation=45,ha='right')
plt.ylabel('Number of jobs')
plt.show()
```
