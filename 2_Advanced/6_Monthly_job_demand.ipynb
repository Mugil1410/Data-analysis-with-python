```python
from datasets import load_dataset
import pandas as pd
import matplotlib.pyplot as plt

#load data
dataSet = load_dataset('lukebarousse/data_jobs')
df = dataSet['train'].to_pandas()


#clean data
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
df_india = df[df['job_country']=='India'].copy()
df_india
```

```python
df_india['job_posted_month'] = df_india['job_posted_date'].dt.strftime('%B')
df_india
```

```python
df_india_pivot=df_india.pivot_table(
    index='job_posted_month',
    columns='job_title_short',
    aggfunc='size',
)
df_india_pivot.reset_index(inplace=True)
df_india_pivot
```

```python
df_india_pivot['job_month_no'] = pd.to_datetime(df_india_pivot['job_posted_month'],format='%B').dt.month
df_india_pivot['job_month_no'] 

df_india_pivot.sort_values(by='job_month_no', inplace=True)
```

```python
df_india_pivot.set_index('job_posted_month',inplace=True)
```

```python
df_india_pivot.drop(columns='job_month_no',inplace=True)
```

```python
df_india_pivot
```

```python
top3 = df['job_title_short'].value_counts().head(3)
top3 = top3.index.to_list()
top3
```

```python
df_india_pivot[top3].plot(kind='line')
plt.title('Monthly Job Postings for Top Data Jobs in the IND')
plt.xlabel('2023')
plt.ylabel('Job Count')
plt.legend()
plt.show()
```
